Use `X,Y=np.meshgrid(x,y)` to evaluate the expression $z=x^2-y^2$ on some grid over the square $(x,y)\in [-2,2]\times [2,2]$.

Plot the graphs $x\mapsto z(x,0)$, $y\mapsto z(0,y)$. 

```python
# CODE HERE
x = np.arange(-2.0, 2.0, 0.1)
y = np.arange(-2.0, 2.0, 0.1)
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, x**2)
axes[1].plot(y, y**2)
axes[0].set_title('$x\\mapsto z(x,0)$')
axes[1].set_title('$y\\mapsto z(0,y)$')
X,Y=np.meshgrid(x,y)
Z = X**2-Y**2
```

Make colormaps and contour figures for the function $z=x^2-y^2$, $(x,y)\in [-2,2]\times [2,2]$.

```python
# CODE HERE
fig, axes = plt.subplots(1,2, figsize=(12, 6))
p = axes[0].pcolor(X, Y, Z, cmap=mpl.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=axes[0])
cnt = axes[1].contour(Z, cmap=mpl.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[-2.0,2.0,-2.0,2.0])
```

Consider four 1000-dimensional samples from the following distributions (with `np.random.seed(42)`)

1. Normal with parameters $(\mu,\sigma^2)$ (`np.random.randn`, `mu=1`, `sigma=1`)
2. Poisson with parameter $\lambda$ (`np.random.poisson`, `lam=1`) 
3. Binomial with parameters $(p,n)$ (`np.random.binomial`, `p=0.6`, `n=10`)
4. Log-normal with parameters $(\mu,\sigma^2)$ (`np.random.lognormal`, `mean=1`, `sigma=1`)

Consider the figure
`fig, axes = plt.subplots(2, 2, figsize=(10, 6))` and make 4 histograms for the mentioned distributions (use 20 `bins`). Give them titles. Use `tight_layout` method.

```python
# CODE HERE
np.random.seed(42)
mu=1
sigma=1
lam=2
p=0.6
n=10
a = sigma * np.random.randn(1000) + mu
b = np.random.poisson(lam, 1000)
c = np.random.binomial(n, p, 1000)
d = np.random.lognormal(mu, sigma, 1000)
```

```python
# CODE HERE
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes[0,0].hist(a,bins = 20)
axes[0,0].set_title("Normal distribution $N(1,1)$")
axes[0,0].set_xlim((min(a), max(a)))
axes[0,1].hist(b,bins = 20)
axes[0,1].set_title("Poisson distribution $\Pi(2)$")
axes[0,1].set_xlim((min(b), max(b)))
axes[1,0].hist(c,bins = 20)
axes[1,0].set_title("Binomial distribution $B(10,0.6)$")
axes[1,0].set_xlim((min(c), max(c)))
axes[1,1].hist(d,bins = 20)
axes[1,1].set_title("Lognormal distribution $LN(1,1)$")
axes[1,1].set_xlim((min(d), max(d)))
```

Download the Red Wine dataset 

https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv

in the pandas dataframe object `red_wine`. Make the scatter plot for 'fixed acidity', 'volatile acidity' columns. Give it a title, give the names for coordinate axes.

```python
# CODE HERE
red_wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
red_wine = pd.read_csv(red_wine_url,sep= ";")
fixed_acidity = red_wine['fixed acidity']
volatile_acidity =  red_wine['volatile acidity']
```

```python
df = red_wine
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(x = df['fixed acidity'], y = df['volatile acidity'])
plt.title('Scatter plot for the first two feature of the Red Wine dataset')
plt.xlabel('volatile acidity')
plt.ylabel('volatile acidity')
plt.show()
```
