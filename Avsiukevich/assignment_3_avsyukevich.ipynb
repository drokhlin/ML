```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Plot the graph of $\sin(x)$ on the interval $[0,100]$. Give the names to the coordinate axes ($x$, $y$). Add the title and the legend.

```python
# CODE HERE
x = np.arange(0,100)
y = np.sin(x)
plt.figure(figsize=(8, 6))
plt.plot(x,y)
plt.ylabel('y')
plt.xlabel('x')
plt.title('The graph of sin')
plt.legend(['sin(x)'])

plt.show()
```

Plot the graphs of $e^x$ (in green), $x^2$ (in red) on the single graph, $x\in [0,2]$. Add the dashed blue grid with "thin" lines. Select the STIX fonts: `mpl.rcParams.update({'font.size': 14, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})`

```python
# CODE HERE
x = np.arange(0, 2, 0.01)
y1 = np.exp(x)
y2 = x**2
mpl.rcParams.update({'font.size': 14, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
plt.figure(figsize=(8, 6))
plt.plot(x, y1, "g-", x, y2, "r-")
plt.ylabel('y')
plt.xlabel('x')
plt.title('The graph of sin')
plt.legend(['exp(x)', '$x^2$'])
plt.grid(color='b', linestyle='--')

plt.show()
```

Use `X,Y=np.meshgrid(x,y)` to evaluate the expression $z=x^2-y^2$ on some grid over the square $(x,y)\in [-2,2]\times [2,2]$.

Plot the graphs $x\mapsto z(x,0)$, $y\mapsto z(0,y)$. 

```python
# CODE HERE
x = np.arange(-2, 2, 0.01)
y = np.arange(-2, 2, 0.01)

fig, (plt1, plt2) = plt.subplots(1, 2)
plt1.plot(x, x**2)
plt2.plot(-y, -y**2)
plt1.set_title('$x\mapsto z(x,0)$')
plt2.set_title('$y\mapsto z(0,y)$')

plt.show(plt1, plt2)
```

Make colormaps and contour figures for the function $z=x^2-y^2$, $(x,y)\in [-2,2]\times [2,2]$.

```python
# CODE HERE
x = np.arange(-2, 2, 0.01)
y = np.arange(-2, 2, 0.01)
X,Y=np.meshgrid(x, y)
z = X**2-Y**2

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.pcolormesh(z)
plt.colorbar()

plt.subplot(1, 2, 2)
plt.contour(z)
plt.show()
```

Consider four 1000-dimensional samples from the following distributions (with `np.random.seed(42)`)

1. Normal with parameters $(\mu,\sigma^2)$ (`np.random.randn`, `mu=1`, `sigma=1`)
2. Poisson with parameter $\lambda$ (`np.random.poisson`, `lam=1`) 
3. Binomial with parameters $(p,n)$ (`np.random.binomial`, `p=0.6`, `n=10`)
4. Log-normal with parameters $(\mu,\sigma^2)$ (`np.random.lognormal`, `mean=1`, `sigma=1`)

Consider the figure
`fig, axes = plt.subplots(2, 2, figsize=(10, 6))` and make 4 histograms for the mentioned distributions (use 20 `bins`). Give them titles. Use `tight_layout` method.

```python
# CODE HERE
np.random.seed(42)
mu=1
sigma=1
lam=2
p=0.6
n=10
```

```python
# CODE HERE
N_points = 1000
n_bins = 20

fig, axes = plt.subplots(2, 2, figsize=(10, 6), tight_layout=True)

axes[0,0].set_title('Normal distribution N(1,1)')
axes[0,1].set_title('Poisson distribution (2)')
axes[1,0].set_title('Binomial distribution B(10,0.6)')
axes[1,1].set_title('Lognormal distribution LN(1,1)')

data = sigma * np.random.randn(N_points) + mu
axes[0,0].hist(data, bins=n_bins)

data = np.random.poisson(lam, N_points)
axes[0,1].hist(data, bins=n_bins)

data = np.random.binomial(n, p, N_points)
axes[1,0].hist(data, bins=n_bins)

data = np.random.lognormal(sigma, mu, N_points)
axes[1,1].hist(data, bins=n_bins)

plt.show()

#   "".  ,   .
```
