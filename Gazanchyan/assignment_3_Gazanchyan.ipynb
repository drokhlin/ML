```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Plot the graph of $\sin(x)$ on the interval $[0,100]$. Give the names to the coordinate axes ($x$, $y$). Add the title and the legend.

```python
# CODE HERE
mpl.rcParams.update({'font.size': 14, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})

x = np.linspace(0, 100, 100)
y = np.sin(x)

fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y, label="$\\sin(x)$") 
axes.legend(loc = 0)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('The graph of $\\sin$')
```

Consider four 1000-dimensional samples from the following distributions (with `np.random.seed(42)`)

1. Normal with parameters $(\mu,\sigma^2)$ (`np.random.randn`, `mu=1`, `sigma=1`)
2. Poisson with parameter $\lambda$ (`np.random.poisson`, `lam=1`) 
3. Binomial with parameters $(p,n)$ (`np.random.binomial`, `p=0.6`, `n=10`)
4. Log-normal with parameters $(\mu,\sigma^2)$ (`np.random.lognormal`, `mean=1`, `sigma=1`)

Consider the figure
`fig, axes = plt.subplots(2, 2, figsize=(10, 6))` and make 4 histograms for the mentioned distributions (use 20 `bins`). Give them titles. Use `tight_layout` method.

```python
# CODE HERE
np.random.seed(42)
mu=1
sigma=1
lam=2
p=0.6
n=10

hist_1 = sigma * np.random.randn(1000) + mu
hist_2 = np.random.poisson(lam, 1000)
hist_3 = np.random.binomial(n, p, 1000)
hist_4 = np.random.lognormal(mu, sigma, 1000)
```

```python
# CODE HERE
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

axes[0,0].hist(hist_1)
axes[0,0].set_title("Normal distribution $N(1,1)$")
axes[0,0].set_xlim((min(hist_1), max(hist_1)))

axes[0,1].hist(hist_2)
axes[0,1].set_title("Poisson distribution $\Pi(2)$")
axes[0,1].set_xlim((min(hist_2), max(hist_2)))

axes[1,0].hist(hist_3)
axes[1,0].set_title("Binomial distribution $B(10,0.6)$")
axes[1,0].set_xlim((min(hist_3), max(hist_3)))

axes[1,1].hist(hist_4)
axes[1,1].set_title("Lognormal distribution $LN(1,1)$")
axes[1,1].set_xlim((min(hist_4), max(hist_4)))

plt.tight_layout()
plt.show()
```

Download the Red Wine dataset 

https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv

in the pandas dataframe object `red_wine`. Make the scatter plot for 'fixed acidity', 'volatile acidity' columns. Give it a title, give the names for coordinate axes.

```python
# CODE HERE
red_wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
red_wine = pd.read_csv(red_wine_url, delimiter=';')

fixed_acidity = red_wine['fixed acidity']
volatile_acidity =  red_wine['volatile acidity']

fig, axes = plt.subplots(1, 1, figsize=(8,6))
axes.scatter(fixed_acidity, volatile_acidity)
axes.set_title("Scatter plot for the first two feature of the Red Wine dataset")
axes.set_xlabel('fixed acidity')
axes.set_ylabel('volatile acidity')
```

For the same columns create the histograms (20 `bins`) on a single figure.

```python
# CODE HERE
fig, axes = plt.subplots(1, 2, figsize=(12,6))

axes[0].hist(fixed_acidity, bins = 20)
axes[0].set_title("Default histogram")
axes[0].set_xlim((min(fixed_acidity), max(fixed_acidity)))
axes[0].set_title("fixed acidity")

axes[1].hist(volatile_acidity, bins=20)
axes[1].set_title("Default histogram")
axes[1].set_xlim((min(volatile_acidity), max(volatile_acidity)))
axes[1].set_title("volatile acidity")
```
