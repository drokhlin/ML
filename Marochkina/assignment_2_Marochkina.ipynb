**Add the column `quality_label`. The values in this column sholud be mapped from `quality` by the following rules:**

`quality_label`=`low` if `quality` is 3,4 or 5

`quality_label`=`medium` if `quality` is 6 or 7

`quality_label`=`high` if `quality` is 8 or 9

**Write the function `qual_transform`, which performs this operation and use `apply` method. Show the first 5 lines of the white wine dataset.**

```python
# CODE HERE
red_wine['quality_label']=np.nan
white_wine['quality_label']=np.nan
```

```python
def qual_transform(a):
  if a in [3,4,5]:
    return 'low'
  elif a in [6,7]:
    return 'medium'
  else:
    return 'high'

    
    
```

```python
# CODE HERE
red_wine['quality_label']=red_wine['quality'].apply(qual_transform)
white_wine['quality_label']=white_wine['quality'].apply(qual_transform)
red_wine.head()
```
