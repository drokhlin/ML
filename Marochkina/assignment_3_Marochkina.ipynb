```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Plot the graph of $\sin(x)$ on the interval $[0,100]$. Give the names to the coordinate axes ($x$, $y$). Add the title and the legend.

```python
# CODE HERE
mpl.rcParams.update({'font.size': 14, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
plt.figure(figsize=(7.7,5))
x = np.linspace(0,100,101)
y = np.sin(x)
plt.plot(x,y, label = 'sin(x)')
ax = plt.gca()
ax.set_title('The graph of sin')
ax.legend(loc='center right')
ax.set_ylabel('y')
ax.set_xlabel('x')
plt.show()
```

Plot the graphs of $e^x$ (in green), $x^2$ (in red) on the single graph, $x\in [0,2]$. Add the dashed blue grid with "thin" lines. Select the STIX fonts: `mpl.rcParams.update({'font.size': 14, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})`

```python
# CODE HERE
mpl.rcParams.update({'font.size': 14, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
plt.figure(figsize=(7.7,5))
x = np.linspace(0,2,101)
y1 = np.exp(x)
y2 = x**2
#plt.subplots(figsize=(7.7,5))
plt.plot(x,y2,color='r', label = '$x^2$')
plt.plot(x,y1,color='g', label = '$e^x$')
ax = plt.gca()
ax.set_title('The graph of $x^2$ and $e^x$')
ax.legend(loc='upper left')
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.grid(color = 'b', linestyle = 'dashed', linewidth = '0.5')
plt.show()
```

Use `X,Y=np.meshgrid(x,y)` to evaluate the expression $z=x^2-y^2$ on some grid over the square $(x,y)\in [-2,2]\times [2,2]$.

Plot the graphs $x\mapsto z(x,0)$, $y\mapsto z(0,y)$. 

```python
# CODE HERE
plt.subplots(1,2)
x = np.linspace(-2,2,101)
y = np.linspace(-2,2,101)
#--
plt.subplot(1,2,1)
plt.plot(x, x**2)
plt.gca().set_title('$x\\mapsto z(x,0)$')
#--
plt.subplot(1,2,2)
plt.plot(y, -y**2)
plt.gca().set_title('$y\\mapsto z(0,y)$')
plt.show()

X,Y = np.meshgrid(x,y)
Z = X**2 - Y**2
```

Make colormaps and contour figures for the function $z=x^2-y^2$, $(x,y)\in [-2,2]\times [2,2]$.

```python
# CODE HERE
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
ax1 = plt.gca()
p = plt.pcolor(X,Y,Z, cmap = mpl.cm.RdBu, vmin= abs(Z).min(), vmax= abs(Z).max())
plt.colorbar(p, ax=ax1)
plt.subplot(1,2,2)
ax2 =plt.gca()
plt.contour(Z, cmap = mpl.cm.RdBu, vmin = abs(Z).min(),vmax = abs(Z).max(), extent=[-2,2,-2,2])
plt.show()
```

Consider four 1000-dimensional samples from the following distributions (with `np.random.seed(42)`)

1. Normal with parameters $(\mu,\sigma^2)$ (`np.random.randn`, `mu=1`, `sigma=1`)
2. Poisson with parameter $\lambda$ (`np.random.poisson`, `lam=1`) 
3. Binomial with parameters $(p,n)$ (`np.random.binomial`, `p=0.6`, `n=10`)
4. Log-normal with parameters $(\mu,\sigma^2)$ (`np.random.lognormal`, `mean=1`, `sigma=1`)

Consider the figure
`fig, axes = plt.subplots(2, 2, figsize=(10, 6))` and make 4 histograms for the mentioned distributions (use 20 `bins`). Give them titles. Use `tight_layout` method.

```python
# CODE HERE
mu=1
sigma=1
lam=2
p=0.6
n=10
```

```python
# CODE HERE
np.random.seed(42)
distr1 = sigma*np.random.randn(1000) + mu
distr2 = np.random.poisson(lam,1000)
distr3 = np.random.binomial(n,p,1000)
distr4 = np.random.lognormal(mu,sigma,1000)

plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.hist(distr1)
plt.gca().set_title('Normal distribution $N(1,1)$')
plt.tight_layout()
#--
plt.subplot(2,2,2)
plt.hist(distr2)
plt.gca().set_title('Poisson distribution $\Pi(2)$')
plt.tight_layout()
#--
plt.subplot(2,2,3)
plt.hist(distr3)
plt.gca().set_title('Binomial distribution $B(10,0.6)$')
plt.tight_layout()
#--
plt.subplot(2,2,4)
plt.hist(distr4)
plt.gca().set_title('Lognormal distribution $LN(1,1)$')
plt.tight_layout()
plt.show()
```

Download the Red Wine dataset 

https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv

in the pandas dataframe object `red_wine`. Make the scatter plot for 'fixed acidity', 'volatile acidity' columns. Give it a title, give the names for coordinate axes.

```python
# CODE HERE
red_wine_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
red_wine = pd.read_csv(red_wine_url, delimiter=';')
fixed_acidity = red_wine['fixed acidity']
volatile_acidity =  red_wine['volatile acidity']
```

```python
plt.subplots(figsize=(7.7,5))
plt.gca().scatter(fixed_acidity,volatile_acidity)
plt.gca().set_title('Scatter plot for the first two feature of the Red Wine dataset')
plt.gca().set_xlabel('fixed acidity')
plt.gca().set_ylabel('volatile acidity')
plt.show()
```

For the same columns create the histograms (20 `bins`) on a single figure.

```python
# CODE HERE
plt.figure(figsize=(10,5))
#--
plt.subplot(1, 2, 1)
plt.gca().hist(fixed_acidity, bins=20)
plt.gca().set_title("fixed acidity")
#--
plt.subplot(1, 2, 2)
plt.gca().hist(volatile_acidity, bins=20)
plt.gca().set_title("volatile acidity")
```
