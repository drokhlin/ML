**Add the column `wine_type` containig the labels `red` and `white` to the corresponding datasets. Show the first 3 lines of the new white wine dataset.**

**Add the column `quality_label`. The values in this column sholud be mapped from `quality` by the following rules:**

`quality_label`=`low` if `quality` is 3,4 or 5

`quality_label`=`medium` if `quality` is 6 or 7

`quality_label`=`high` if `quality` is 8 or 9

**Write the function `qual_transform`, which performs this operation and use `apply` method. Show the first 5 lines of the white wine dataset.**

```python
# CODE HERE
def qual_transform(a):
    if a['quality'] in [3,4,5]:
        a['quality_label'] = 'low'
    
    if a['quality'] in [6,7]:
        a['quality_label'] = 'medium'
    
    if a['quality'] in [8,9]:
        a['quality_label'] = 'high'
        
    return(a)
```

```python
# CODE HERE
red_wine = red_wine.apply(qual_transform,axis = 1)
red_wine.head()
```

```python
# CODE HERE
white_wine = white_wine.apply(qual_transform,axis = 1)
white_wine.head()
```

**Concatenate `red_wine` and `white_wine` into `wine`. Show its first 5 lines.**

```python
# CODE HERE
wine = pn.concat([red_wine,white_wine])
wine.head()
```
