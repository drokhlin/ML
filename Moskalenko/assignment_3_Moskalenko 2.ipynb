```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Plot the graph of $\sin(x)$ on the interval $[0,100]$. Give the names to the coordinate axes ($x$, $y$). Add the title and the legend.

```python
# CODE HERE
mpl.rcParams.update({'font.size': 14, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
x = np.linspace(0,100,100)
figure1 = plt.figure().add_axes([0,0,1,1])
figure1.plot(np.sin(x),label = '$sin(x)$')
figure1.set_xlabel('x')
figure1.set_ylabel('y')
figure1.set_title('The graph of sin')
figure1.legend(loc = 5)
```

Plot the graphs of $e^x$ (in green), $x^2$ (in red) on the single graph, $x\in [0,2]$. Add the dashed blue grid with "thin" lines. Select the STIX fonts: `mpl.rcParams.update({'font.size': 14, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})`

```python
# CODE HERE
x = np.linspace(0,2,10)
figure2 = plt.figure().add_axes([0,0,1,1])
figure2.plot(x,np.exp(x),color = 'green',label = '$exp(x)$')
figure2.plot(x,x**2,color = 'red',label = '$x^2$')
figure2.set_xlabel('x')
figure2.set_ylabel('y')
figure2.set_title('The graphs of $e^x$ and $x^2$')
figure2.grid(color = 'blue', alpha = 1, linestyle = 'dashed', linewidth=0.5)
figure2.legend(loc = 2)
```

Use `X,Y=np.meshgrid(x,y)` to evaluate the expression $z=x^2-y^2$ on some grid over the square $(x,y)\in [-2,2]\times [2,2]$.

Plot the graphs $x\mapsto z(x,0)$, $y\mapsto z(0,y)$. 

```python
# CODE HERE
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)
fig,axes = plt.subplots(1,2)
axes[0].plot(x,x**2)
axes[0].set_title('$x\\mapsto z(x,0)$')
axes[1].plot(y,-y**2)
axes[1].set_title('$y\\mapsto z(0,y)$')

X,Y = np.meshgrid(x,y)
Z = X**2-Y**2
```

Make colormaps and contour figures for the function $z=x^2-y^2$, $(x,y)\in [-2,2]\times [2,2]$.

```python
# CODE HERE
fig,axes = plt.subplots(1,2,figsize = (12,6))
p = axes[0].pcolor(X, Y, Z, cmap=mpl.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(),shading = 'auto')
cb = fig.colorbar(p, ax=axes[0])
cnt = axes[1].contour(Z, cmap=mpl.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(),extent = [-2,2,-2,2])
```

Consider four 1000-dimensional samples from the following distributions (with `np.random.seed(42)`)

1. Normal with parameters $(\mu,\sigma^2)$ (`np.random.randn`, `mu=1`, `sigma=1`)
2. Poisson with parameter $\lambda$ (`np.random.poisson`, `lam=1`) 
3. Binomial with parameters $(p,n)$ (`np.random.binomial`, `p=0.6`, `n=10`)
4. Log-normal with parameters $(\mu,\sigma^2)$ (`np.random.lognormal`, `mean=1`, `sigma=1`)

Consider the figure
`fig, axes = plt.subplots(2, 2, figsize=(10, 6))` and make 4 histograms for the mentioned distributions (use 20 `bins`). Give them titles. Use `tight_layout` method.

```python
# CODE HERE
np.random.seed(42)
mu=1
sigma=1
lam=1 # Your hist has lam = 2
p=0.6
n=10
g1 = sigma*np.random.randn(1000)+mu
g2 = np.random.poisson(lam,1000)
g3 = np.random.binomial(n,p,1000)
g4 = np.random.lognormal(mu,sigma,1000)
```

```python
# CODE HERE
fig, axes = plt.subplots(2,2,figsize = (12,8))
axes[0,0].set_title('Normal distribution $N(1,1)$')
axes[0,0].hist(g1,bins = 20) # Your hists have bins = 10
axes[0,0].set_xlim((min(g1),max(g1)))

axes[0,1].set_title('Poisson distribution $\Pi(1)$')
axes[0,1].hist(g2,bins = 20)
axes[0,1].set_xlim((min(g2),max(g2)))

axes[1,0].set_title('Binomial distribution $B(10, 0.6)$')
axes[1,0].hist(g3,bins = 20)
axes[1,0].set_xlim((min(g3),max(g3)))

axes[1,1].set_title('Lognormal distribution $LN(1,1)$')
axes[1,1].hist(g4,bins = 20)
axes[1,1].set_xlim((min(g4),max(g4)))
```

Download the Red Wine dataset 

https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv

in the pandas dataframe object `red_wine`. Make the scatter plot for 'fixed acidity', 'volatile acidity' columns. Give it a title, give the names for coordinate axes.

```python
# CODE HERE
red_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', 
                       delimiter = ';')
fig, axes = plt.subplots(figsize = (8,6))
axes.scatter(red_wine['fixed acidity'], red_wine['volatile acidity'])
axes.set_title('Scatter plot for the first two feature of the Red Wine dataset')
axes.set_xlabel('fixed acidity')
axes.set_ylabel('volatile acidity')
```

For the same columns create the histograms (20 `bins`) on a single figure.

```python
# CODE HERE
g1 = red_wine['fixed acidity']
g2 = red_wine['volatile acidity']
fig,axes = plt.subplots(1,2,figsize = (10,5))
axes[0].set_title('fixed acidity')
axes[0].hist(g1,bins = 20)
axes[0].set_xlim((min(g1),max(g1)))
axes[1].set_title('volatile acidity')
axes[1].hist(g2,bins = 20)
axes[1].set_xlim((min(g2),max(g2)))
```
